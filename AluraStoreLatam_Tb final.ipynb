#1. Análisis de facturación



```python
def suma_x_tienda(td):
    return td['Precio'].sum()

suma_t1 = suma_x_tienda(tienda)
suma_t2 = suma_x_tienda(tienda2)
suma_t3 = suma_x_tienda(tienda3)
suma_t4 = suma_x_tienda(tienda4)

print('Ingreso total tienda 1: ', int(suma_t1))
print('Ingreso total tienda 2: ', int(suma_t2))
print('Ingreso total tienda 3: ', int(suma_t3))
print('Ingreso total tienda 4: ', int(suma_t4))
```

```python
import matplotlib.pyplot as plt

# Suma de los datos con las ventas de cada tienda
ingresos = [suma_t1, suma_t2, suma_t3, suma_t4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(8, 5))
plt.bar(tiendas, ingresos, color='skyblue')
plt.title('Ingresos por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, valor in enumerate(ingresos):
    plt.text(i, valor + max(ingresos)*0.01, f"${int(valor)}", ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
def conteo_categorias(df, top_n=5):
    conteo = df['Categoría del Producto'].value_counts().head(top_n)
    print(conteo)
    return conteo

def lista_categorias_tienda(df):
    categorias = df['Categoría del Producto'].dropna().unique().tolist()
    print(categorias)
    return categorias

tiendas = [tienda, tienda2, tienda3, tienda4]

for i, t in enumerate(tiendas, start=1):
    print(f"Categorías de Tienda {i}:")
    lista_categorias_tienda(t)
    print()
```

```python
def categorias_tiendas(td):
    return list(td['Categoría del Producto'].unique())

categorias_T1 = categorias_tiendas(tienda)
categorias_T2 = categorias_tiendas(tienda2)
categorias_T3 = categorias_tiendas(tienda3)
categorias_T4 = categorias_tiendas(tienda4)

print(categorias_T1)
print(categorias_T2)
print(categorias_T3)
print(categorias_T4)

def cantidad_x_categorias(tienda):
    return tienda['Categoría del Producto'].value_counts().to_dict()

cant_t1 = cantidad_x_categorias(tienda)
cant_t2 = cantidad_x_categorias(tienda2)
cant_t3 = cantidad_x_categorias(tienda3)
cant_t4 = cantidad_x_categorias(tienda4)

print(cant_t1)
print(cant_t2)
print(cant_t3)
print(cant_t4)
```

```python
import matplotlib.pyplot as plt

# Conteos por tienda
conteos_por_tienda = [cant_t1, cant_t2, cant_t3, cant_t4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Loop graficando las tiendas por separado
for i, conteo in enumerate(conteos_por_tienda):
    plt.figure(figsize=(6, 6))
    plt.pie(
        conteo.values(),
        labels=conteo.keys(),
        autopct='%1.1f%%',
        startangle=90,
        colors=plt.cm.Set3.colors
    )
    plt.title(f'Distribución por Categoría - {nombres_tiendas[i]}', fontsize=13)
    plt.axis('equal')
    plt.tight_layout()
    plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Identificamos valores nulos en las tiendas

print(tienda['Calificación'].isnull().sum())
print(tienda2['Calificación'].isnull().sum())
print(tienda3['Calificación'].isnull().sum())
print(tienda4['Calificación'].isnull().sum())
```

```python
def calificacion_promedio(tienda):
    return tienda['Calificación'].mean()

calif_t1 = calificacion_promedio(tienda)
calif_t2 = calificacion_promedio(tienda2)
calif_t3 = calificacion_promedio(tienda3)
calif_t4 = calificacion_promedio(tienda4)

print(f"Calificación promedio Tienda 1: {calif_t1:.2f}")
print(f"Calificación promedio Tienda 2: {calif_t2:.2f}")
print(f"Calificación promedio Tienda 3: {calif_t3:.2f}")
print(f"Calificación promedio Tienda 4: {calif_t4:.2f}")
```

```python
import matplotlib.pyplot as plt

calificaciones = [calif_t1, calif_t2, calif_t3, calif_t4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(8, 5))
barras = plt.bar(tiendas, calificaciones, color='lightcoral')

for i, barra in enumerate(barras):
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width() / 2, altura + 0.1, f'{calificaciones[i]:.2f}',
             ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.title('Calificación Promedio por Tienda')
plt.ylabel('Calificación (de 1 a 5)')
plt.ylim(0, 5.5)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
def productos_mas_y_menos_vendidos(tienda):
    conteo_productos = tienda['Producto'].value_counts()

    productos_top = conteo_productos.head(5)

    productos_bottom = conteo_productos.tail(5)

    return productos_top, productos_bottom
```

```python
productos_top_t1, productos_bottom_t1 = productos_mas_y_menos_vendidos(tienda)
productos_top_t2, productos_bottom_t2 = productos_mas_y_menos_vendidos(tienda2)
productos_top_t3, productos_bottom_t3 = productos_mas_y_menos_vendidos(tienda3)
productos_top_t4, productos_bottom_t4 = productos_mas_y_menos_vendidos(tienda4)

# Mostrando los resultados por cada tienda

print("Tienda 1 - Productos más vendidos:")
print(productos_top_t1)
print("Tienda 1 - Productos menos vendidos:")
print(productos_bottom_t1)

print("Tienda 2 - Productos más vendidos:")
print(productos_top_t2)
print("Tienda 2 - Productos menos vendidos:")
print(productos_bottom_t2)

print("Tienda 3 - Productos más vendidos:")
print(productos_top_t1)
print("Tienda 3 - Productos menos vendidos:")
print(productos_bottom_t1)

print("Tienda 4 - Productos más vendidos:")
print(productos_top_t2)
print("Tienda 4 - Productos menos vendidos:")
print(productos_bottom_t2)
```

```python
def resumen_productos_mas_y_menos_vendidos(productos_top, productos_bottom, tienda_nombre):
    if isinstance(productos_top, pd.Series):
        productos_top = productos_top.items()
    if isinstance(productos_bottom, pd.Series):
        productos_bottom = productos_bottom.items()

    # Verificación de que exista al menos un proucto
    if not productos_top or not productos_bottom:
        print(f"Error: No hay productos disponibles en Tienda {tienda_nombre}")
        return

    try:

        prod_max, cant_max = list(productos_top)[0]
        prod_min, cant_min = list(productos_bottom)[-1]

        print(f"Resumen Tienda {tienda_nombre}:")
        print(f"- Producto más vendido: {prod_max} con {cant_max} unidades")
        print(f"- Producto menos vendido: {prod_min} con {cant_min} unidades\n")

    except Exception as e:
        print(f"Error al procesar Tienda {tienda_nombre}: {e}")

# Resumen de cada tienda
resumen_productos_mas_y_menos_vendidos(productos_top_t1, productos_bottom_t1, '1')
resumen_productos_mas_y_menos_vendidos(productos_top_t2, productos_bottom_t2, '2')
resumen_productos_mas_y_menos_vendidos(productos_top_t3, productos_bottom_t3, '3')
resumen_productos_mas_y_menos_vendidos(productos_top_t4, productos_bottom_t4, '4')
```

# 5. Envío promedio por tienda

```python
def promedio_costo_envio(envio):
    return envio['Costo de envío'].mean()

envios_tienda = promedio_costo_envio(tienda)
envios_tienda2 = promedio_costo_envio(tienda2)
envios_tienda3 = promedio_costo_envio(tienda3)
envios_tienda4 = promedio_costo_envio(tienda4)

print(f'Promedio costo de envío tienda 1: {envios_tienda:.2f}')
print(f'Promedio costo de envío tienda 2: {envios_tienda2:.2f}')
print(f'Promedio costo de envío tienda 3: {envios_tienda3:.2f}')
print(f'Promedio costo de envío tienda 4: {envios_tienda4:.2f}')
```

```python
import matplotlib.pyplot as plt

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
promedios_envio = [envios_tienda, envios_tienda2, envios_tienda3, envios_tienda4]

plt.plot(tiendas, promedios_envio, marker='o', color='green', linestyle='-', label='Costo de Envío')

plt.title('Promedio Costo de Envío por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de Envío (en $)')

plt.grid(True)

plt.legend()

plt.tight_layout()
plt.show()
```
